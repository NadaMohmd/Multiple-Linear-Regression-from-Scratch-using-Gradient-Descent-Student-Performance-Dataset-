```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"/content/sample_data/Student_Performance.csv"
student_performance = pd.read_csv(path)
```

# Encode Categorical Column

```python
student_performance['Extracurricular Activities'] = student_performance['Extracurricular Activities'].map({'No': 0, 'Yes': 1})
```

# Z-score normalization

```python
student_performance = (student_performance - student_performance.mean()) / student_performance.std()
```

#  Add Bias Term (Intercept Column)

```python
student_performance.insert(0, 'ones', 1)
```

```python
student_performance.describe()
```

#  Separate Features and Target Variable

```python
x = student_performance.drop('Performance Index', axis = 1)
y = student_performance['Performance Index']

# cols = student_performance.shape[1]
# x =student_performance.iloc[ : ,0 :cols-1]
# y =student_performance.iloc[ : ,cols-1 :cols]
```

```python
# Convert to matrix and define theta
x = np.matrix(x.values)
y = np.matrix(y.values)
theta = np.matrix(np.zeros((1, x.shape[1])))
```

#  Define Cost Function

```python
def computeCost(x , y , theta):
    z = np.power(((x * theta.T)- y), 2)
    return np.sum(z) / (2 * len(x))
print("cost = ", computeCost(x , y , theta))
```

# Implement Vectorized Gradient Descent

```python
def computeGradient_vectorized(x , y , theta , alpha , iteras):
    cost = np.zeros(iteras)

    for i in range(iteras):
        error = x * theta.T - y
        gradient = (alpha / len(x)) * (x.T * error)
        theta = theta - gradient.T
        cost[i] = computeCost(x , y , theta)

    return theta, cost
alpha = 0.01
iteras = 500
g , cost = computeGradient_vectorized(x , y , theta , alpha , iteras)
```

```python
print('g = ', g)
print('cost = ', cost[0:50])
print('computeCost = ',computeCost(x , y , g))
```
